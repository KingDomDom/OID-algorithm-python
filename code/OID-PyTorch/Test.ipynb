```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 创建示例数据
data = {
    '': ['1', '1', '2', '2', '3', '3', '4', '4'],
    'Type': ['OID', 'OID w/o Coarse Deblur', 'OID', 'OID w/o Coarse Deblur', 'OID', 'OID w/o Coarse Deblur', 'OID', 'OID w/o Coarse Deblur'],
    'PSNR': [34.7671, 31.4082, 32.5998, 31.2601, 35.8071, 34.6323, 30.8549, 28.9946]
}

df = pd.DataFrame(data)

# 创建条形图
plt.figure(figsize=(5, 4))
sns.barplot(x='', y='PSNR', hue='Type', data=df)
plt.ylim(20, 40)

# 添加标题和标签
plt.ylabel('PSNR')

# 显示图形
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 创建示例数据
data = {
    '': ['1', '1', '2', '2', '3', '3', '4', '4'],
    'Type': ['OID', 'OID w/o Coarse Deblur', 'OID', 'OID w/o Coarse Deblur', 'OID', 'OID w/o Coarse Deblur', 'OID', 'OID w/o Coarse Deblur'],
    'SSIM': [0.9424, 0.8880, 0.8940, 0.8638, 0.9078, 0.8824, 0.8874, 0.8432]
}

df = pd.DataFrame(data)

# 创建条形图
plt.figure(figsize=(5, 4))
sns.barplot(x='', y='SSIM', hue='Type', data=df)
plt.ylim(0.5, 1)

# 添加标题和标签
plt.ylabel('SSIM')

# 显示图形
plt.show()
```
